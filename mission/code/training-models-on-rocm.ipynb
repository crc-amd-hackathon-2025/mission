```python
import wandb
from huggingface_hub import login
wandb.login(key="<wandb_token>")
login(token="<hf_token>")
```

```python
from huggingface_hub import HfApi

#Just to make sure.
hub_api = HfApi()
hub_api.delete_tag("crc-amd-hackathon-2025/grab-cam", tag="v3.0", repo_type="dataset")
hub_api.create_tag("crc-amd-hackathon-2025/grab-cam", tag="v3.0", repo_type="dataset")
```

```python
!lerobot-train \
    --dataset.repo_id=crc-amd-hackathon-2025/grab-cam \
    --policy.type=pi05 \
    --output_dir=./outputs/crc-amd-hackathon-2025/pi05-grab-cam-3 \
    --job_name=pi05-grab-cam-3 \
    --policy.repo_id=crc-amd-hackathon-2025/pi05-grab-cam-3\
    --policy.pretrained_path=lerobot/pi05_base \
    --policy.compile_model=true \
    --policy.gradient_checkpointing=true \
    --wandb.enable=true \
    --policy.dtype=bfloat16 \
    --steps=5000 \
    --policy.device=cuda \
    --batch_size=32 \
    --policy.push_to_hub=true
```

```python
!hf upload crc-amd-hackathon-2025/pi05-grab-cam-3 \
  ./outputs/crc-amd-hackathon-2025/pi05-grab-cam-3/checkpoints/last/pretrained_model
```
